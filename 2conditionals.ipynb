```python
if age == 18:
    print("You are eligible to vote")
```

```python
# elif statement
# allows you to check another condition if the previous if was false

# elif condition:
#     code

# else = if all the if's and elif's are false then this will happen


# else:
#     code
```

```python
age = 22
citizen = True  
if citizen and age >= 18:
    print("You are eligible to vote")
elif citizen and age < 18:
    print("You are a citizen but not old enough to vote")
else:
    print("You are not a citizen, so you cannot vote")
```

```python
# multiple conditions in in single if

age = 25
citizen = True
criminal_record = False

if age >= 18 and citizen and not criminal_record:
    print("You are eligible to vote")
else:
    print("You are not eligible to vote")
```
