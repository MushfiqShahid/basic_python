```python
# for loop

# 1. basic use for range

# for variable in range(start, stop, step):
#     # code to execute
```

```python
# prints only the even numbers between 2 and 20
for variable in range(2, 20, 2):
    print(variable)

```

```python
# 2 loopint through a list

fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

```python
# 3 Looping through a string
for char in "Python":
    print(char)
```

```python
# 4 looping through a dictionary

person = {"name": "Alice", "age": 25}

# keys
for key in person:
    print(key, "->", person[key])

# values
for value in person.values():
    print(value)

# items
for key, value in person.items():
    print(key, value)
```

```python
# 5 nested for loop

for i in range(1, 4):          # outer loop
    for j in range(1, 4):      # inner loop
        print(i, j)
```

```python
# 6 using and end statement 
for i in range(1, 5):
    print(i)
else:
    print("Loop completed!")
```

```python
# 7 List comprehensions = compacted for loop
squares = [x**20 for x in range(1, 100)]
print(squares)
```

```python
# 8 iterating over multiple sequences using zip
names = ["Alice", "Bob", "Charlie"]
ages = [25, 30, 22]

for name, age in zip(names, ages):
    print(name, "is", age, "years old")
```

```python
# 9 using for with enumerate

fruits = ["apple", "banana", "cherry"]

for index, fruit in enumerate(fruits):
    print(index, fruit)
```

```python
# 10 break and continue
for num in range(1, 11):
    if num % 3 == 0:
        continue  # skip numbers divisible by 3
    if num > 8:
        break     # stop the loop if number > 8
    print(num)
```

```python
rows = int(input("Enter how many rows of Pascal's Triangle you want to print: "))

for i in range(rows):
    num = 1
    print(" " * (rows - i), end="")

    for j in range(i + 1):
        print(num, end=" ")
        num = num * (i - j) // (j + 1) 

    print() 
```
