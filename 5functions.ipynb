```python
# keyword arguments
def student(name, age):
    print(f"{name} is {age} years old")

student(age=24, name="Mushfiq")
```

```python
# *args = you don't know how many arguments you will pass then use this

def add_all(*numbers):
    total = 0
    for n in numbers:
        total += n
    return total

print(add_all(1, 2, 3, 4, 5, 6, 7, 8)) 
```

```python
# **kwargs = when you want to use key value pairs of data

def user_info(**info):
    for key, value in info.items():
        print(f"{key}: {value}")

user_info(name="Mushfiq", age=24, city="Dhaka")
```

```python
# nested function

def outer():
    def inner():
        print("This is inner function")
    print("This is outer function")
    inner()

outer()
```

```python
# returns multiple values
def get_student():
    return "Mushfiq", 24, "Dhaka"

name, age, city = get_student()
print(name, age, city)
```

```python
# decorator = this is a type of special function which takes other functions as argument
# this type of function is imposed upon another function to take that 2nd function as and input argument for the decorator
```

```python
# a decorator function for checking runtime of another function

import time

def timing(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        runtime = end_time - start_time
        print(f"{func.__name__} took {runtime:.10f} seconds to run")
        return result
    return wrapper
```

```python
# here we can see the timing decorator is being imposed on pring_pascal_triangle and showing the  running time of the function
@timing
def print_pascals_triangle(rows):
    """Print Pascal's Triangle up to the given number of rows."""
    for i in range(rows):
        num = 1
        print(" " * (rows - i), end="")

        for j in range(i + 1):
            print(num, end=" ")
            num = num * (i - j) // (j + 1)

        print()  # move to next line


# ---- Main Program ----
rows = int(input("Enter how many rows of Pascal's Triangle you want to print: "))
print_pascals_triangle(rows)
```
