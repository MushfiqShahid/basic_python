```python
# while loop

# while condition:
#     code
#     incrementation
```

```python
count = 1

while count <= 5:
    print("Count is:", count)
    count += 1   # increase count each time
```

```python
# nested while loop

# while condition1:
#     # outer loop code
#     while condition2:
#         # inner loop code
```

```python
# 2 directional counting
i = 1
while i <= 3:         # outer loop
    j = 1
    while j <= 2:     # inner loop
        print("i =", i, "j =", j)
        j += 1
    i += 1
```

```python
# printing a 3x3 matrix using while loop

i = 1         # row counter
num = 1       # number to print

while i <= 3:   # outer loop for rows
    j = 1       # column counter
    while j <= 3:  # inner loop for columns
        print(num, end=" ")  # print number on the same line
        num += 1
        j += 1
    print()  # move to next line after inner loop
    i += 1
```

```python
# infinite loop
while True:
    print("This will run forever")
```

```python
# break = stops the loop on condition is met
# continue = skips that iteration and continues the rest on condition
while True:
    num = int(input("Enter a number (0 to stop): "))

    if num < 0:
        print("Negative numbers are ignored!")
        continue  # skip the rest and ask again

    if num == 0:
        print("Exiting the program...")
        break     # stop the loop

    print("You entered:", num)
```
